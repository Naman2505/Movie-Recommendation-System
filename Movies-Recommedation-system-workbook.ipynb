## Modelling

Now data is preprocessed now we have to do text vectorization where we convert text into vectors.
method for doing so:

### 1️⃣ Basic Method

1) Bag of Words-
* Converts text into a frequency-based representation.
* Ignores the order of words.
* 

``` 
from sklearn.feature_extraction.text import CountVectorizer

corpus = ["I love NLP and AI", "AI loves me"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())  # Vocabulary
print(X.toarray())  # Vectorized text
```

2) TF-IDF (Term Frequency-Inverse Document Frequency)
* Adjusts word frequency by penalizing common words.
* Helps reduce bias towards frequently occurring words.
* 

```
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = ["I love NLP and AI", "AI loves me"]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())

```

### 2️⃣ Word Embeddings (Dense Representations)
Unlike BoW or TF-IDF, embeddings capture meaning and relationships between words.

1) Word2Vec (Using gensim)
* Converts words into dense vectors based on their usage in large corpora.
* Two approaches: CBOW (Continuous Bag of Words) and Skip-gram.
```
from gensim.models import Word2Vec

sentences = [["I", "love", "NLP"], ["AI", "loves", "me"], ["Machine", "learning", "is", "fun"]]
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

print(model.wv["love"])  # Vector for "love"
```

2) FastText (Word2Vec but better)
* Captures subword information (useful for handling rare words or misspellings).
* Example: "running" = ["run", "unning"] → better generalization.
```
from gensim.models import FastText

model = FastText(sentences, vector_size=100, window=5, min_count=1, workers=4)
print(model.wv["love"])  # Vector for "love"
```
3) GloVe (Global Vectors)
* Pre-trained embeddings trained on word co-occurrence.
* Available for different corpus sizes (e.g., Wikipedia, Common Crawl).
```
import gensim.downloader as api

glove_vectors = api.load("glove-wiki-gigaword-100")  # Load pre-trained GloVe 100D vectors
print(glove_vectors["love"])  # Vector for "love"
```
### 3️⃣ Contextualized Embeddings (Deep Learning)
Unlike Word2Vec, these models generate different vector representations for words based on context.

1) BERT (Bidirectional Encoder Representations from Transformers)
* Uses transformers to capture context dynamically.
* Produces different vectors for "bank" in:
"I deposited money in the bank."
"I sat near the river bank."
```
from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

text = "I love NLP and AI"
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)

print(outputs.last_hidden_state.shape)  # Token-wise embeddings
```
### 🚀 Summary: Best Approach for Your Use Case

| Method | Use Case |
|:----------:|:----------:|
| BoW    | Simple ML models, small datasets   |
| TF-IDF    | Document classification, keyword extraction   |
| Word2Vec/FastText	|NLP tasks like sentiment analysis recommendation|
|GloVe|	General NLP tasks pre-trained vector usage|
|BERT	|Context-sensitive NLP (chatbots, question answering)|

## 1️⃣ Tokenization (Splitting Text)
Tokenization breaks text into words or sentences.

```
from nltk.tokenize import word_tokenize, sent_tokenize

text = "I love NLP! It's amazing."
print(word_tokenize(text))  # ['I', 'love', 'NLP', '!', "It's", 'amazing', '.']
print(sent_tokenize(text))  # ['I love NLP!', "It's amazing."]
```
🔹 Use Case: Preparing text for NLP models.

## 2️⃣ Removing Stopwords
Stopwords are common words like "the", "is", "and", which add little meaning.

```
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
words = word_tokenize("This is an amazing NLP tutorial!")
filtered_words = [word for word in words if word.lower() not in stop_words]

print(filtered_words)  # ['amazing', 'NLP', 'tutorial', '!']
```
🔹 Use Case: Reducing noise in text data.

## 3️⃣ Stemming (Reducing Words to Their Root)
Stemming reduces words to their base form.

```
from nltk.stem import PorterStemmer

ps = PorterStemmer()
print(ps.stem("running"))  # run
print(ps.stem("flies"))  # fli
```
🔹 Use Case: Search engines & text normalization.

## 4️⃣ Lemmatization (Better than Stemming)
Lemmatization reduces words to their base form using dictionary meanings.

```
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running", pos="v"))  # run
print(lemmatizer.lemmatize("better", pos="a"))  # good
```
🔹 Use Case: More accurate NLP models.

## 5️⃣ Part-of-Speech (POS) Tagging
Assigns grammatical categories (noun, verb, adjective, etc.).

```
from nltk import pos_tag

words = word_tokenize("I love deep learning.")
nltk.download('averaged_perceptron_tagger')
print(pos_tag(words))

Output:
[('I', 'PRP'), ('love', 'VBP'), ('deep', 'JJ'), ('learning', 'NN'), ('.', '.')]
```
🔹 Use Case: Named Entity Recognition (NER), grammar checking.

## 6️⃣ Named Entity Recognition (NER)
Identifies names, places, dates, and other entities.

```
from nltk.chunk import ne_chunk

nltk.download('maxent_ne_chunker')
nltk.download('words')

sentence = "Elon Musk founded SpaceX in California."
words = word_tokenize(sentence)
tags = pos_tag(words)
ner_tree = ne_chunk(tags)

print(ner_tree)
```
🔹 Use Case: Extracting company names, people, locations.

## 7️⃣ Sentiment Analysis
Classifies text as positive, negative, or neutral.

```
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

print(sia.polarity_scores("I love this product!"))  # {'neg': 0.0, 'neu': 0.25, 'pos': 0.75, 'compound': 0.6369}
print(sia.polarity_scores("This is the worst thing ever!"))  # {'neg': 0.654, 'neu': 0.346, 'pos': 0.0, 'compound': -0.8271}
```
🔹 Use Case: Analyzing reviews, social media sentiment.

## Rock Paper Scissors

````python
def winner(comp,person):
    if(comp == "Rock" and person=="paper"):
        return "You"
    elif (comp == "Rock" and person=="Scissors"):
        return "Computer"
    elif (comp == "paper" and person=="Rocks"):
        return "Computer"
    elif (comp == "paper" and person=="Scissors"):
        return "You"
    elif (comp == "Scissors" and person=="paper"):
        return "Computer"
    elif (comp == "Scissors" and person=="Rock"):
        return "You"
    else:
        return "Tie"

def ShowR(W,L,T):
    print(f"Your Score : wins : {W}, losses : {L}, Ties : {T}\n")

import random

option = {0:"Rock",1:"Paper",2:"Scissors"}

turns = 3

win = 0
loss = 0
tie = 0

ShowR(win,loss,tie)

print(f"You have {turns} games left\n")

while turns:
    
    Person = option[int(input("Enter your Choice (Rock:0, Paper:1,Scissors:2) : "))]
    Computer = option[random.randint(0,2)]
    
    result = winner(Computer,Person)
    
    if result == "You":
        win+=1
    elif result == "Computer":
        loss+=1
    else:
        tie+=1
    turns-=1
    
    ShowR(win,loss,tie)
````
